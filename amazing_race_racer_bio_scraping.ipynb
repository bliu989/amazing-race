```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

# Scraping the data

```python
# scrape teams wiki page for The Amazing Race US
url1 = "https://amazingrace.fandom.com/wiki/Category:The_Amazing_Race_(US)_teams"
req1 = requests.get(url1)
soup1 = bs(req1.content, 'html.parser')

url2 = "https://amazingrace.fandom.com/wiki/Category:The_Amazing_Race_(US)_teams?from=Kishori+%26+Karishma"
req2 = requests.get(url2)
soup2 = bs(req2.content, 'html.parser')
```

```python
teams = soup1.find_all('li', 
                       attrs={'class': 'category-page__member'}) 
teams.extend(soup2.find_all('li', 
                       attrs={'class': 'category-page__member'}))
```

```python
team_names = [team.a['title'] for team in teams]
```

```python
url_head = 'https://amazingrace.fandom.com/'
```

```python
team_urls = [url_head + team.a['href'] for team in teams]
```

```python
team_bios = []

for url in team_urls:
    print(url)
    bio = ""

    req = requests.get(url)
    soup = bs(req.content, 'html.parser')

    profile_header = soup.find('span', attrs={'id': 'Profile'})
    if profile_header is None:
        profile_header = soup.find('span', attrs={'id': 'Profiles'})
    if profile_header is None:
        profile_header = soup.find('span', attrs={'id': 'CBS_Biography'})
    if profile_header is None:
        profile_header = soup.find('span', attrs={'id': 'Race_History'})


    for tag in profile_header.parent.next_siblings:
        if tag.name == 'h2':
            break
        else:
            s = re.sub(r"(.*?<b>)(?:.*?)(</b>.*)", r'\1\2', str(tag))
            s = re.sub(r"(.*?<i>)(?:.*?)(</i>.*)", r'\1\2', s)
            bio += bs(s).text
    
    team_bios.append(bio)
```

```python
df = pd.DataFrame({
    'team_name': team_names,
    'team_bio': team_bios
})
```

```python
df.to_csv('data/bios.csv', index=False)
```
